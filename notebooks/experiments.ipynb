```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("Healthcare Data Cleaning").master("local[*]").getOrCreate()

# Path to the CSV file
csv_file_path = r'C:\Users\vamsh\OneDrive\Desktop\Healthcare_data\healthcare-data-pipeline\data\01_raw\patients.csv'

# Load the CSV file into a DataFrame
patient_df = spark.read.csv(
    csv_file_path,
    sep=',',  # Specify the separator if necessary (comma in this case)
    header=True,  # Use the first row as column names
    inferSchema=True  # Automatically infer data types
)

# Column renaming
column_mapping = {
    'PATIENT_ID': 'patient_id',
    'BIRTHDATE': 'birth_date',
    'DEATHDATE': 'death_date',
    'SSN': 'ssn',
    'DRIVERS': 'drivers_license',
    'PASSPORT': 'passport_number',
    'PREFIX': 'name_prefix',
    'FIRST': 'first_name',
    'LAST': 'last_name',
    'SUFFIX': 'name_suffix',
    'MAIDEN': 'maiden_name',
    'MARITAL': 'marital_status',
    'RACE': 'race_ethnicity',
    'GENDER': 'gender',
    'BIRTHPLACE': 'birthplace',
    'ADDRESS': 'address',
    'CITY': 'city',
    'STATE': 'state',
    'COUNTY': 'county',
    'FIPS': 'fips_code',
    'ZIP': 'zip_code',
    'LAT': 'latitude',
    'LON': 'longitude',
    'HEALTHCARE_EXPENSES': 'healthcare_expenses',
    'HEALTHCARE_COVERAGE': 'healthcare_coverage',
    'INCOME': 'annual_income'
}

# Renaming columns in Spark DataFrame
for old_col, new_col in column_mapping.items():
    if old_col in patient_df.columns:
        patient_df = patient_df.withColumnRenamed(old_col, new_col)

# Show the first few rows and the schema to verify
patient_df.show(5) 
```
